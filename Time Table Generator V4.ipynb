```python
def create_population(size):
    population = []
    for i in range(size):
        schedule = {}
        for dept in departments:
            schedule[dept] = {}
            for subject in departments[dept]['subjects']:
                total_classes = departments[dept]['total_classes'][departments[dept]['subjects'].index(subject)]
                professors = departments[dept]['professors'][subject]
                schedule[dept][subject] = {
                    'total_classes': total_classes,
                    'professor': random.choice(professors),
                    'room': random.choice(rooms),
                    'days': random.sample([i for i in range(len(days))], total_classes),
                    'timeslots': random.sample([i for i in range(len(timeslots))], total_classes)
                }
        population.append(schedule)
    return population
create_population(3)
```

```python
# Define a function to calculate the fitness of a schedule
def calculate_fitness(schedule):
    conflicts = 0
    for dept in schedule:
        for subject in schedule[dept]:
            classes = schedule[dept][subject]['days']
            for i in range(len(classes)):
                for j in range(i+1, len(classes)):
                    if classes[i] == classes[j]:
                        if schedule[dept][subject]['timeslots'][i] == schedule[dept][subject]['timeslots'][j]:
                            conflicts += 1
                        #elif schedule[dept][subject]['room'][i] == schedule[dept][subject]['room'][j]:
                        #    conflicts += 1
                        elif schedule[dept][subject]['professor'][i] == schedule[dept][subject]['professor'][j]:
                            conflicts += 1
    return 1 / (conflicts + 1)
x = [calculate_fitness(timetable) for timetable in create_population(3)]
x
```

```python
def select_parents(population):
    fitnesses = [calculate_fitness(schedule) for schedule in population]
    total_fitness = sum(fitnesses)
    probabilities = [fitness / total_fitness for fitness in fitnesses]
    parents = random.choices(population, weights=probabilities, k=2)
    return parents
```

```python
def crossover(parent1, parent2):
    child = {}
    for dept in parent1:
        child[dept] = {}
        for subject in parent1[dept]:
            total_classes = parent1[dept][subject]['total_classes']
            professor1 = parent1[dept][subject]['professor']
            professor2 = parent2[dept][subject]['professor']
            if professor1 == professor2:
                professor = professor1
            else:
                professor = random.choice([professor1, professor2])
            room1 = parent1[dept][subject]['room']
            room2 = parent2[dept][subject]['room']
            if room1 == room2:
                room = room1
            else:
                room = random.choice([room1, room2])
            days1 = parent1[dept][subject]['days']
            days2 = parent2[dept][subject]['days']
            timeslots1 = parent1[dept][subject]['timeslots']
            timeslots2 = parent2[dept][subject]['timeslots']
            days = []
            timeslots = []
            for i in range(total_classes):
                if days1[i] == days2[i]:
                    days.append(days1[i])
                    if timeslots1[i] == timeslots2[i]:
                        timeslots.append(timeslots1[i])
                    else:
                        timeslots.append(random.choice([timeslots1[i], timeslots2[i]]))
                else:
                    days.append(random.choice([days1[i], days2[i]]))
                    timeslots.append(random.choice([timeslots1[i], timeslots2[i]]))
            child[dept][subject] = {
                'total_classes': total_classes,
                'professor': professor,
                'room': room,
                'days': days,
                'timeslots': timeslots
            }
    return child
```

```python
# Define a function to mutate a schedule
def mutate(schedule, mutation_rate):
    for dept in schedule:
        for subject in schedule[dept]:
            if random.random() < mutation_rate:
                total_classes = schedule[dept][subject]['total_classes']
                schedule[dept][subject]['professor'] = random.choice(departments[dept]['professors'][subject])
                schedule[dept][subject]['room'] = random.choice(rooms)
                schedule[dept][subject]['days'] = random.sample([i for i in range(len(days))], total_classes)
                schedule[dept][subject]['timeslots'] = random.sample([i for i in range(len(timeslots))], total_classes)
    return schedule
```

```python
def next_generation(current_population, elite_size, mutation_rate):
    elite = sorted(current_population, key=lambda x: calculate_fitness(x), reverse=True)[:elite_size]
    selection_size = len(current_population) - elite_size
    parents = [select_parents(current_population) for i in range(selection_size)]
    children = []
    for parent1, parent2 in parents:
        child = crossover(parent1, parent2)
        if random.random() < mutation_rate:
            child = mutate(child, 0.1)
        children.append(child)
    next_generation = elite + children
    return next_generation
```

```python
# Genetic algorithm
def genetic_algorithm(population_size, mutation_rate, generations):
    # Initialize population
    population = create_population(population_size)
    for generation in range(generations):
        # Calculate fitness scores
        #x = [calculate_fitness(timetable) for timetable in create_population(3)]
        
        fitness_scores = [calculate_fitness(timetable) for timetable in population]
        
        # Select parents
        parents = [select_parents(population) for i in range(population_size)]
        
        # Crossover
        children = [crossover(parents[i][0], parents[i][1]) for i in range(population_size)]
        
        # Mutation
        mutated_children = [mutate(child, mutation_rate) for child in children]
        
        # Create new population
        population = mutated_children
        
        # Print best timetable in each generation
        best_timetable = max(population, key=calculate_fitness)
        print(f"Generation {generation + 1}: Best fitness score = {calculate_fitness(best_timetable)}, Best timetable = {best_timetable}")
    
    return best_timetable
```

```python
# Example usage
best_timetable = genetic_algorithm(population_size=10, mutation_rate=0.1, generations=10)
print(f"Final timetable: {best_timetable}")
```
